# DickensAssignment2

This program will inspect and transform the MV Rankin Data and meta to get validated for Nature's Palette repository submission.

```python
import os
import pandas as pd
import numpy as np
```

```python
df_meta = pd.read_csv('metadata.csv')
df_meta.head()
```

### Extensions

The first issue to be noticed is the file extension is not included in the FileName.

```python
df_meta_output = df_meta.copy()
df_meta_output['FileName'] = df_meta['FileName'] + ".txt"
df_meta_output.head()
```

### Duplicates

Check for duplicate rows

```python
print('Original number of rows', df_meta.shape[0])
df_meta_output = df_meta_output.reset_index().drop_duplicates(subset='FileName', keep='first').reset_index()
df_meta_output = df_meta_output.dropna(subset=['Replicate'])
print('After dropping duplicates', df_meta_output.shape[0])
df_meta_output
```

### Look for non-matches

```python
# copy the filenames from the directory
filenames = [name for path, subdirs, files in os.walk("RankinData")
             for name in files]

df_filenames = pd.DataFrame (filenames, columns = ['filename'])
df_filenames
```

```python
df_missing_meta = df_filenames[~df_filenames.filename.isin(df_meta_output.FileName.values)]

print("Number of files without meta present: ", df_missing_meta.shape[0])

# export files without meta
df_missing_meta.to_csv('ValidatorExports/MissingMeta.csv', index=False, header=True)
```

```python
df_missing_files = df_meta_output[~df_meta_output.FileName.isin(filenames)]
df_missing_files.drop(['index', 'level_0'], axis=1, inplace=True)

print("Number of meta without files present: ", df_missing_files.shape[0])

# export meta without files
df_missing_files.to_csv('ValidatorExports/MissingFiles.csv', index=False, header=True)
```
